## **PRINCIPAL COMPONENT ANALYSIS**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
pcadata=pd.read_csv(r"C:\Users\hp\Desktop\iris.csv")
print(pcadata)
```

```python
x=pcadata[["sepal_length","sepal_width","petal_length","petal_width"]]
print(x)
y=pcadata["species"]
print(y)
```

```python
std_x=StandardScaler().fit_transform(x)
std_x=pd.DataFrame(std_x)
print(std_x)
```

```python
print(std_x.mean().round(6))
print(std_x.std(ddof=0).round(6))
```

```python
pca=PCA(n_components=2)
pca_model=pca.fit_transform(std_x)
pca_model
```

```python
pca.explained_variance_ratio_
```

```python
pcadf=pd.DataFrame(pca_model,columns=["pc1","pc2"])
pcadf= pd.concat([pcadf,pcadata["species"]],axis=1)
pcadf
```
