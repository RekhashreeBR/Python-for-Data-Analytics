## **CONTINUOUS PROBABILITY DISTRIBUTION**

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from scipy.stats import uniform, expon, gamma, beta
```

```python
a = 0  
b = 20  
x = np.linspace(a - 2, b + 2, 1500)
pdf_values = uniform.pdf(x, loc=a, scale=b-a)
plt.plot(x, pdf_values, color='GREEN', label=f"Uniform({a}, {b})")
plt.fill_between(x, pdf_values, alpha=0.3, color='PURPLE')
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.title("Continuous Uniform Distribution PDF")
plt.legend()
plt.ylim(0, 0.2)
plt.show()
```

```python
lambda_param = 1 
x = np.linspace(0, 5, 1000)
pdf = expon.pdf(x, scale=1/lambda_param)
plt.plot(x, pdf, label=f'Exponential Distribution\nLambda={lambda_param}')
plt.title('Exponential Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
shape = 2 
scale = 2 
x = np.linspace(0, 20, 1000)
pdf = gamma.pdf(x, a=shape, scale=scale)
pdf
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label=f'Gamma PDF (shape={shape}, scale={scale})', color='RED')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Gamma Probability Distribution')
plt.legend()
plt.grid()
```

```python
alpha, beta_param = 2, 50 
x = np.linspace(0, 1, 5000)

pdf_values = beta.pdf(x, alpha, beta_param)
plt.plot(x, pdf_values, color='BLACK', label=f"Beta({alpha}, {beta_param})")
plt.fill_between(x, pdf_values, alpha=0.3, color='red')
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.title("Beta Distribution PDF")
plt.legend()
plt.show()
```
