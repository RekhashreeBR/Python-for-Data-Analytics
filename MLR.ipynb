## **MULTIPLE LINEAR REGTRESSION**

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x=DATA34[["Volume","Weight"]]
y=DATA34["CO2"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80, random_state=0)
```

```python
mlr_model=LinearRegression()
mlr_model.fit(x_train,y_train)
```

```python
y_pred=mlr_model.predict(x_test)
y_pred
```

```python
y_test
```

```python
print("intercept= ", mlr_model.intercept_)
print("coeffecient =", mlr_model.coef_)
print("mse =", mean_squared_error(y_test,y_pred))
print("r2_scoree =", r2_score(y_test,y_pred))
```

```python

pred_DATA34=pd.DataFrame({"actual":y_test,"predicted":y_pred,"difference":y_test-y_pred})
pred_DATA34
```
