## **DISCRETE PROBABILITY DISTRIBUTION**

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from scipy.stats import binom, geom, poisson, bernoulli, hypergeom
```

```python
p=0.8                  
x =[0,1]

pmf_values = bernoulli.pmf(x,p) 
pmf_values
plt.bar(x, pmf_values, alpha=0.8)
plt.xticks(x)
plt.xlabel("Success/Failure")
plt.ylabel("Probability")
plt.title(f"Bernoulli Distribution (p = {p})")
plt.ylim(0, 1)
plt.show()
```

```python
n = 20 
p = 0.5
x = np.arange(0, n+1)
binom_pmf = binom.pmf(x, n, p)
print("Probability of Getting x Heads in 10 Flips:")
print(dict(zip(x, binom_pmf)))
x_binom = range(len(binom_pmf)) 
plt.bar(x_binom, binom_pmf, color="PURPLE")
plt.title("Coin Flips")
plt.xlabel("Number of Heads in 10 Flips")
plt.ylabel("Probability")
plt.show()
```

```python
p = 0.5 
x = np.arange(1, 20) 
geom_pmf = geom.pmf(x, p)
print("Probability of First Heads on x-th Flip:")
print(dict(zip(x, geom_pmf)))
x_geom = range(len(geom_pmf)) 
plt.bar(x_geom, geom_pmf, color="BLUE")
plt.title("Geometric Distribution - First Heads")
plt.xlabel("Trial Number")
plt.ylabel("Probability")
plt.show()
```

```python
N = 20  
K = 7   
n = 5   
x = np.arange(0, n+1)

pmf_values = hypergeom.pmf(x, N, K, n)
pmf_values
plt.bar(x, pmf_values, color='ReD', alpha=0.7)
plt.xlabel("Number of Successes (k)")
plt.ylabel("Probability P(X = k)")
plt.title(f"Hypergeometric Distribution (N={N}, K={K}, n={n})")
plt.show()
```

```python
lambda_ = 1             
x = np.arange(0, 10)    
poisson_pmf = poisson.pmf(x, lambda_)
print(" Probability of x Rare Events Occurring:")
print(dict(zip(x, poisson_pmf)))
x_poisson = range(len(poisson_pmf))      
plt.bar(x_poisson, poisson_pmf, color="BLACK")
plt.title("Poisson Distribution - Rare Events")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.show()
```
